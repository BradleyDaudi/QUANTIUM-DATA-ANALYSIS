```python
#dropping null values
purch.dropna(inplace = True)
```

```python
#renaming the 'PREMIUM_CUSTOMER' to 'CUSTOMER TYPE'
purch.rename(columns={'PREMIUM_CUSTOMER': 'CUSTOMER TYPE'}, inplace=True)
```

```python
#we can rseparate the values in 'LIFESTAGE' removing the '/'
purch[['LIFESTAGE', 'SPLIT']] = purch['LIFESTAGE'].str.split('/',expand = True)
```

```python
#we therefore remove the Column 'SPLIT' for we dont need it
purch.drop(columns = 'SPLIT', inplace = True)
```

```python
#Creating a funvtion to read the second dataset
def readat1(file_path):
    df2 = pd.read_csv(file_path)
    
    
    return df2
```

```python
#loading the dataframe
df2 = readat1("C:\\Users\\USER\\Desktop\\quantium data analysis project\\Untitled Folder/QVI_transaction_data.csv")
```

```python
#making a copy of the dataset
transc = df2.copy()
```

```python
#display the data
transc.head(60)
```

```python
#checking for null values
transc.info()
transc.isnull().sum()
```

```python
# Convert 'DATE' column to 'yyyy-mm-dd' format
transc['DATE'] = pd.to_datetime(transc['DATE'], origin='1899-12-30', unit='D').dt.strftime('%Y-%m-%d')
```

```python
#display to see date change
transc.head(20)
```

```python
# change the 'PROD_NAME' to  the same case
transc['PROD_NAME'] = transc['PROD_NAME'].str.lower()
```

```python
transc.head(20)
```

```python
#separating the string and the digits to be in the same formating as the rest of the trings (with a ' ' between the string and the digit containg the grams)
import re

transc['PROD_NAME'] = transc['PROD_NAME'].apply(lambda x: re.sub(r'(?<=[a-zA-Z])(?=\d)', ' ', x))

```

```python
#display to see the changes
transc.head()
```

```python
#splitting the column 'PROD_NAME'
# Extract the quantity information using regular expressions
transc['QTY-GRAMS'] = transc['PROD_NAME'].str.extract(r'(\d+g)')
transc.head()
```

```python
#removing the digit + 'g' in the 'PROD_NAME'
transc['PROD_NAME'] = transc['PROD_NAME'].str[:-4].str.strip()
# Print the modified dataframe
transc.head()
```

```python
#joining the first dataset with the second data set using the primary key
merged_data = pd.merge(left=purch, right=transc, on='LYLTY_CARD_NBR', how='right')

```

```python
#check the total number of rows and columns after merging in comparison to the others
print('merged_data = ', merged_data.shape)
print('purch =', purch.shape)
print('transc =', transc.shape)
```

```python
#save the merged data set to csv
merged_data.to_csv('merged_dataset.csv', index=False)
```
